```python
#INTRODUCTION

Cybersecurity has become a critical concern in today’s digital world due to the rapid increase in cyber attacks and data breaches. 
This project analyzes cybersecurity attack data to identify patterns, trends, and potential risks using data visualization techniques. 
    
Various aspects such as attack types, severity levels, network protocols, malware indicators, and anomaly scores are examined to understand how attacks occur and how severe they are. 
The insights gained from this analysis help in improving threat detection, strengthening security measures, and supporting better decision-making in cybersecurity management.
```

```python
import pandas as pd

#Load Dataset

df = pd.read_csv(r"C:\Users\abhin\Documents\jupyter stuff\cybersecurity_attacks.csv")
df
```

```python
# Display the last 5 rows of the dataset
df.tail()
```

```python
# Check duplicate Columns
df.columns.duplicated()
```

```python
# Check duplicate Rows
df.duplicated().sum()
```

```python
# Checking and removing duplicate rows to maintain data integrity
df = df.drop_duplicates()
```

```python
# View dataset information
df.info()
```

```python
# Check null values 
df.isnull().sum()
```

```python
# Drop rows that contain null (missing) values
df.drop(['IDS/IPS Alerts', 'Firewall Logs', 'Proxy Information', 'Alerts/Warnings'], 
        axis=1, inplace=True, errors='ignore')
```

```python
# Check null values 
df.isnull().sum()
```

```python
# Filling Missing Values
df['Malware Indicators'] = df['Malware Indicators'].fillna(0)
```

```python
# Check null values 
df.isnull().sum()
```

```python
print("Missing values handled successfully")
```

```python
# This shows statistical summary of dataset
df.describe()
```

```python
# Visualization
```

```python
# Shows the most frequent types of cyber attacks, helping identify the most common threats.

import matplotlib.pyplot as plt

df['Attack Type'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Cyber Security Attacks")
plt.xlabel("Attack Type")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Displays how attacks are distributed across severity levels (Low, Medium, High), indicating overall risk levels.

df['Severity Level'].value_counts().plot(kind='bar')

plt.title("Severity Level Distribution")
plt.xlabel("Severity Level")
plt.ylabel("Count")
plt.show()
```

```python
# Identifies which network protocols are most commonly targeted or exploited in attacks.

df['Protocol'].value_counts().plot(kind='bar')

plt.title("Network Protocol Usage in Attacks")
plt.xlabel("Protocol")
plt.ylabel("Count")
plt.show()
```

```python
# Highlights the systems or sources where attacks are detected, helping improve monitoring strategies.

df['Log Source'].value_counts().plot(kind='bar')

plt.title("Source of Attack Logs")
plt.xlabel("Log Source")
plt.ylabel("Count")
plt.show()
```

```python
# Shows the relationship between malware presence and attack severity, indicating how malware impacts risk.

pd.crosstab(df['Malware Indicators'], df['Severity Level']).plot(kind='bar', stacked=True)

plt.title("Malware Indicators vs Severity Level")
plt.xlabel("Malware Indicator")
plt.ylabel("Count")
plt.show()
```

```python
# Represents how unusual or suspicious each attack type is based on anomaly detection metrics.

df.groupby('Attack Type')['Anomaly Scores'].mean().sort_values().plot(kind='bar')

plt.title("Average Anomaly Score by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Confirms consistency in anomaly behavior across different attack categories.

avg_anomaly = df.groupby('Attack Type')['Anomaly Scores'].mean().sort_values()

avg_anomaly.plot()

plt.title("Average Anomaly Score by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Average Anomaly Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
